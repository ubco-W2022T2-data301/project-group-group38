```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('http://data.insideairbnb.com/united-states/hi/hawaii/2022-12-15/data/listings.csv.gz')
df
```

```python
df['price'] = df['price'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
Island_Names = df['neighbourhood_group_cleansed'].unique()
Island_Names
```

Now I can see that there is information for these 4 islands. My initial research question is "Which island in the Hawaiian islands has the lowest average for an airbnb rental.

To answer this I need the average price for a rental on each island.

```python
avg_price = df.groupby('neighbourhood_group_cleansed')['price'].mean()
avg_price = pd.DataFrame({'Island': avg_price.index, 'Average Price': avg_price.values})
avg_price
```

Now let's visualize the data.

```python
ax1 = avg_price.plot(kind='bar')
ax1.set_title('Average Price for airbnb by Island')
ax1.set_xlabel('Island')
ax1.set_ylabel('Price in USD')
```

I can see now that my initial question is too trivial. 

I also see that Maui appears to have much more expensive listings but I also know that this graph does not tell the whole story as there are multiple types of rooms/houses that you can rent and these numbers represent the average of all rentals. Common sense tells us that it would be more expensive to rent an entire house vs. a hotel room or shared room.

I have identified the "room_type" column and will now explore this.

```python
df['room_type'].unique()
```

So now I can see that there are 4 different room types that were listed and included in this dataset. I think it would be useful to know how many of each room types there are on each island.

```python
grouped_df = df.groupby(['neighbourhood_group_cleansed', 'room_type']).size().reset_index(name='count')
grouped_df
```

```python
plt = sns.catplot(x='neighbourhood_group_cleansed', y='count', hue='room_type', data=grouped_df, kind='bar')
plt.fig.suptitle('Number of airbnb listings by Room Type and Island')
plt.set(xlabel='Island', ylabel='Number of listings')
```

Observations:

1. The vast majority of these airbnb listings are Entire home/apt and most of the remainder are Private rooms.
2. Now that I can see that most of the listings fall under these two categories, I am interested in the average price for each of  these room types on each island

I will start by finding the average price of each room type for all 4 islands.

```python
entire_home_df = df[df['room_type'] == 'Entire home/apt']
avg_price_entire_home = entire_home_df['price'].mean()
avg_price_entire_home
```

```python
private_room_df = df[df['room_type'] == 'Private room']
avg_price_Private_room = private_room_df['price'].mean()
avg_price_Private_room
```

This is interesting as it suggests the average price for a Private room is more expensive thatn for an entire home or apartment.
