**Which Island in the Hawaiian islands has the lowest average price for an Airbnb rental?**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('http://data.insideairbnb.com/united-states/hi/hawaii/2022-12-15/data/listings.csv.gz')
df
```

```python
df.describe()
```

```python
df.describe(include=[np.number])
```

```python
df.describe(exclude=[np.number])
```

```python
df['price'] = df['price'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
Island_Names = df['neighbourhood_group_cleansed'].unique()
Island_Names
```

Now I can see that there is information for these 4 islands. My initial research question is "Which island in the Hawaiian islands has the lowest average for an airbnb rental.

To answer this I need the average price for a rental on each island.

```python
avg_price = df.groupby('neighbourhood_group_cleansed')['price'].mean()
avg_price = pd.DataFrame({'Island': avg_price.index, 'Average Price': avg_price.values})
avg_price
```

Now let's visualize the data.

```python
ax1 = avg_price.plot(kind='bar')
ax1.set_title('Average Price for airbnb by Island')
ax1.set_xlabel('Island')
ax1.set_ylabel('Price in USD')
```

I can see now that my initial question is too trivial. 

I also see that Maui appears to have much more expensive listings but I also know that this graph does not tell the whole story as there are multiple types of rooms/houses that you can rent and these numbers represent the average of all rentals. Common sense tells us that it would be more expensive to rent an entire house vs. a hotel room or shared room.

I have identified the "room_type" column and will now explore this.

```python
df['room_type'].unique()
```

So now I can see that there are 4 different room types that were listed and included in this dataset. I think it would be useful to know how many of each room types there are on each island.

```python
grouped_df = df.groupby(['neighbourhood_group_cleansed', 'room_type']).size().reset_index(name='count')
grouped_df
```

```python
plt = sns.catplot(x='neighbourhood_group_cleansed', y='count', hue='room_type', data=grouped_df, kind='bar')
plt.fig.suptitle('Number of airbnb listings by Room Type and Island')
plt.set(xlabel='Island', ylabel='Number of listings')
```

Observations:

1. The vast majority of these airbnb listings are Entire home/apt and most of the remainder are Private rooms.
2. Now that I can see that most of the listings fall under these two categories, I am interested in the average price for each of  these room types on each island

I will start by finding the average price of each room type for all 4 islands.

```python
entire_home_df = df[df['room_type'] == 'Entire home/apt']
avg_price_entire_home = entire_home_df['price'].mean()
avg_price_entire_home
```

```python
private_room_df = df[df['room_type'] == 'Private room']
avg_price_Private_room = private_room_df['price'].mean()
avg_price_Private_room
```

This is interesting as it suggests the average price for a Private room is more expensive thatn for an entire home or apartment.

Now I would like to see what the average prices for these two room types on each of the 4 islands.

```python
avg_price_private_room_by_island = private_room_df.groupby(['neighbourhood_group_cleansed', 'room_type'])['price'].mean()
df_avg_price_private_room_by_island = pd.DataFrame(avg_price_private_room_by_island).reset_index().sort_values(by=['price'], ascending=False)
df_avg_price_private_room_by_island
```

```python
avg_price_by_island = entire_home_df.groupby(['neighbourhood_group_cleansed', 'room_type'])['price'].mean()
df_avg_price_by_island = pd.DataFrame(avg_price_by_island).reset_index().sort_values(by=['price'], ascending=False)
df_avg_price_by_island
```

Let's visualize this data for each room type.

```python
sns.set_style('darkgrid')
colors = {'Honolulu': 'teal', 'Kauai': 'green', 'Maui': 'blue', 'Hawaii': 'purple'}
barplot2 = sns.barplot(x='neighbourhood_group_cleansed', y='price', data=df_avg_price_by_island, palette = colors)
barplot2.set(xlabel='Island', ylabel='Average Price', title='Average Price of Private Room Rentals on Each Island')
```

```python
sns.set_style('whitegrid')
colors = {'Honolulu': 'teal', 'Kauai': 'green', 'Maui': 'blue', 'Hawaii': 'purple'}
barplot3 = sns.barplot(x='neighbourhood_group_cleansed', y='price', data=df_avg_price_private_room_by_island, palette = colors)
barplot3.set(xlabel='Island', ylabel='Average Price', title='Average Price of Entire Home/Apt Rentals on Each Island')
```

Ok, so I can now see each island's average price for the entire home/apt column and the Private room column. 
I want to see if I can combine some of this data to get the full picture.

Price is not everything. Usually a higher price is correlated with higher quality and vice versa. I would like to find out the relation between price and quality using some of the review score data contained in the dataset.

I will start by looking at the average review score for each of the islands.

```python
df = pd.read_csv('http://data.insideairbnb.com/united-states/hi/hawaii/2022-12-15/data/listings.csv.gz')
df = df[['neighbourhood_group_cleansed', 'review_scores_rating']]
df = df.groupby('neighbourhood_group_cleansed').mean()
df
```

As we can see, the review scores are very similar for each island and are generally very well-rated but maybe the scores differ based on the room type.

```python
df = pd.read_csv('http://data.insideairbnb.com/united-states/hi/hawaii/2022-12-15/data/listings.csv.gz')
df = df[['neighbourhood_group_cleansed', 'room_type', 'review_scores_rating']]
df = df[df['room_type'].isin(['Private room', 'Entire home/apt'])]
df = df.groupby(['neighbourhood_group_cleansed', 'room_type']).mean().round(2)
df = df.unstack()
df.columns = df.columns.droplevel()
df.reset_index(inplace=True)
df.rename(columns={'neighbourhood_group_cleansed': 'Island'}, inplace=True)
df.set_index('Island', inplace=True)
df.columns = ['Private Room', 'Entire home/apt']
df
```

```python
sns.set_style("whitegrid")
sns.barplot(x=df.index, y="Private Room", data=df, color="skyblue", label="Private Room")
plt.title("Average Review Scores For Private Rooms by Island in Hawaii")
plt.ylabel("Average Review Scores")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x=df.index, y="Entire home/apt", data=df, color="salmon", label="Entire home/apt")
plt.title("Average Review Scores for Entire home/apt in Hawaii")
plt.ylabel("Average Review Scores")
plt.show()
```

As this table and visualization shows, the scores are still very similar and do not suggest that there is much difference in the quality of the airbnb stay based on either island or room type. 

Now that I have this information, I would like to investigate the relationship between the prices and review scores of both room types for each island.

```python
df = pd.read_csv('http://data.insideairbnb.com/united-states/hi/hawaii/2022-12-15/data/listings.csv.gz')
df = df[['neighbourhood_group_cleansed', 'room_type', 'review_scores_rating']]
df = df[df['room_type'].isin(['Private room', 'Entire home/apt'])]
df = df.groupby(['neighbourhood_group_cleansed', 'room_type']).mean().round(2)
df = df.unstack()
df.columns = df.columns.droplevel()
df.reset_index(inplace=True)
df.rename(columns={'neighbourhood_group_cleansed': 'Island'}, inplace=True)
df.set_index('Island', inplace=True)
df.columns = ['Private Room', 'Entire home/apt']
df
```

Here is the average prices of Private rooms for each island.
